### Problem Statement ###
Given the root of a binary tree, check whether it is a mirror of itself (i.e., symmetric around its center).

<br /> Example 1:



<br />Input: root = [1,2,2,3,4,4,3]
<br />Output: true

<br />Example 2:



<br />Input: root = [1,2,2,null,3,null,3]
<br />Output: false
 

<br />Constraints:
<br />The number of nodes in the tree is in the range [1, 1000].
<br />-100 <= Node.val <= 100

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def checkSymmetry(self, l_node, r_node):
        if l_node == None and r_node == None:
            return True

        if l_node == None or r_node == None:
            return False

        if l_node.val != r_node.val:
            return False

        return (self.checkSymmetry(l_node.left, r_node.right) and 
                self.checkSymmetry(l_node.right, r_node.left))

    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
        return self.checkSymmetry(root.left, root.right)
```

### Complexity Analysis ###

## Time ##

- O(N), where N is the number of nodes in the tree. 
- We look at every node in the tree once

## Space ##

- Worst case: O(N)
    - the tree is unbalanced, so we will have a max of N stack frames at once.
- Average case: O(logN)
    - the tree is balanced, so we will have log N stack frames at once.
